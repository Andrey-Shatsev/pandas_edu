#### Применение loc и iloc

```python
# то, как работает loc:
#coffee.loc[Rows, Columns]

coffee.loc[0] #возвращается серия в виде первой строки 
```

```python
coffee.loc[[0,1,2]] # вернем первые три строки
```

```python
coffee.loc[5:8, ["Day", 'Units Sold']] # выберем только тек колонки которые нужны - к нашим строкам
```

```python
coffee.iloc[:, [0,2]] # все строки и два столбца
```

```python
# можно пользоваться заменами индексов на что-то значащие:
coffee.index = coffee["Day"]
```

```python
coffee.loc[1:3, "Units Sold"] = 10
```

```python
coffee
```

```python
coffee.at[0, "Units Sold"] # получаем значение на пересечении - для единичного значения
```

```python
coffee.sort_values(["Units Sold", 'Coffee Type'], ascending = [0,1]) 
# можно задать разные направления для разных столбцов
```

#### Как происходит итерация:

```python
# переборы - не особо оптимизирован под циклы - но для анализа - ок:
for index, row in coffee.iterrows():
    print(index)
    print(row['Units Sold'])
```

### Фильтрация данных:

```python
bios.info()
```

```python
# отфильтровали строки и взяли нужные нам столбцы:
bios.loc[bios['height_cm']>215, ['name', 'height_cm']].sort_values('height_cm')
```

```python
#можно фильтровать проще:
bios[bios['height_cm']>215][['name', 'height_cm']]
```

```python
#Несколько условий:
bios[(bios['height_cm']>215) & (bios['born_country']=='USA')]
```

```python
# поиск по содержанию: чувствительно к регистру
bios[bios['name'].str.contains('Keith|patrick', case=False)] # можно использовать синтаксис регулярных вражений, его можно отключить - чтобы искать точно совпадение
```

```python
#можно записать запрос без повторения наименования df:
bios.query('born_country == "USA"')
```
