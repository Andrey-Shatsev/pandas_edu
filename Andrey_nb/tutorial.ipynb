```python
bios_new.info()
```

```python
def categorize_athlete(row):
    if pd.isna(row['height_cm']) or pd.isna(row['weight_kg']):
        return 'Unknown'    
    elif row['height_cm'] < 170 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] >= 170 and row['height_cm'] < 190 and row['weight_kg'] >= 70 and row['weight_kg'] < 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'
```

```python
bios_new['Category'] = bios_new.apply(categorize_athlete, axis=1)
```

```python
bios_new[['name', 'height_cm',  'weight_kg', 'Category']].head()
```

### Работа с Null

```python
import numpy as np
```

```python
#coffee = pd.read_csv('../complete-pandas-tutorial/warmup-data/coffee.csv')
#coffee
```

```python
coffee.loc[[0,1], 'Units Sold'] = np.nan
```

```python
#как видим теперь часть данных у нас null:
coffee.info()
```

```python
#чтобы посчитать количество null:
coffee.isna().sum()
```

```python
# что можем сделать:
coffee.fillna(30) # хардом заполнить каким-нибудь значением
```

```python
# более разумным решением будет:
coffee.fillna(coffee['Units Sold'].mean())
```

```python
# еще более мудрое решение - интерполировать. Для этого нужно переставить данные местами
n = len(coffee)
```

```python
idx = np.concatenate([
    [2, 3],              # Кого двигаем
    np.arange(0, 2),     # Кто был в начале
    np.arange(4, n)      # Все остальные
])
```

```python
coffee = coffee.iloc[idx].reset_index(drop=True)
```

```python
coffee['Units Sold'].interpolate()
```

```python
# можно дропнить часть na (смотрим только на subset)
coffee.dropna(subset='Units Sold')
```

```python
# можем отобрать только значения na в конкретном столбце:
coffee[coffee['Units Sold'].isna()] # либо notna()
```

```python
#отобрать nan во всех столбцах:
rows_with_nan = coffee[coffee.isna().any(axis=1)]
rows_with_nan
```

```python
# 1. Создаем список с правильным порядком
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
# pivots:
pivot = coffee.pivot(columns='Coffee Type', index='Day', values='revenue')
pivot
```

```python
#можно фильтровать pivot:
pivot.loc['Monday', 'Latte']
```

```python
# 2. Переопределяем индекс как категориальный с заданным порядком
pivot.index = pd.CategoricalIndex(pivot.index, categories=days_order, ordered=True)

# 3. Сортируем по этому индексу
pivot = pivot.sort_index()
pivot
```

```python
pivot.sum() # тоталы по колонкам
pivot.sum(axis=0)
```

```python
pivot.sum(axis=1) #тоталы по строкам
```
