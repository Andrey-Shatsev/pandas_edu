```python
bios_new = bios.copy()
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new.head()
```

```python
#поработаем с объектами даты и времени
bios_new['born_date']= pd.to_datetime(bios_new['born_date'])
```

```python
bios_new.info()
```

```python
bios_new['Year of Birth'] = bios_new['born_date'].dt.year
```

```python
bios_new[['name', 'Year of Birth']].head()
```

```python
#простые функции с lyambda:
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Tall' if x > 190 else 'Average' if x > 170 else 'Short')
```

```python
bios_new[['name', 'height_cm', 'height_category']].head(10)
```

```python
bios_new.info()
```

```python
def categorize_athlete(row):
    if pd.isna(row['height_cm']) or pd.isna(row['weight_kg']):
        return 'Unknown'    
    elif row['height_cm'] < 170 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] >= 170 and row['height_cm'] < 190 and row['weight_kg'] >= 70 and row['weight_kg'] < 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'
```

```python
bios_new['Category'] = bios_new.apply(categorize_athlete, axis=1)
```

```python
bios_new[['name', 'height_cm',  'weight_kg', 'Category']].head()
```
